NAIVE BAYES (MULTINOMIAL NAVIE BAYES)


```python
import pandas as pd
import numpy as np
import sklearn

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_iris

```

```python
irisds=load_iris()
irisds
```

```python
irisds.keys()
```

```python
irisds.target
```

```python
irisds.feature_names
```

```python
df=pd.DataFrame(data=irisds.data,columns=irisds.feature_names)
df
```

```python
df['target']=irisds.target
```

```python
df
```

```python
df.dtypes
```

```python
df.info()
```

```python
x=df.iloc[:,0:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_train,y_train)
```

```python
predmnb=mnb.predict(x_test)
predmnb
```

```python
print (accuracy_score(y_test,predmnb))
```

```python
print (confusion_matrix(y_test,predmnb))
```

```python
print (classification_report(y_test,predmnb))
```

```python
def irispredict (p):
    p=p.reshape(1,-1)
    predmnb=mnb.predict(p)
    print ('predict value: ',predmnb)
    if predmnb==0:
        print ('Sitosa')
    elif predmnb==1:
        print ('Versiculer')
    elif predmnb==2:
        print ('virginica')
        
    
    
```

```python
p=np.array([6.1,4.5,1.2,0.6])
irispredict (p)
```

NAIVE BAYES (GAUSSIAN NAVIE BAYES)

```python
df2=pd.read_csv('cancerfile.csv')
```

```python
df2
```

```python
x=df2.iloc[:,0:-1]
x
```

```python
y=df2.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
predgnb=gnb.predict(x_test)
predgnb
```

```python
print (accuracy_score(y_test,predgnb))
print (confusion_matrix(y_test,predgnb))
print (classification_report(y_test,predgnb))
```
