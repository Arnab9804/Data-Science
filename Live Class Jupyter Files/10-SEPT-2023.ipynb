# Gradient Descent Algorithm

```python
import numpy as np
import pandas as pd

import sklearn
from sklearn import linear_model
from sklearn.linear_model import SGDRegressor
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.pipeline import Pipeline

from sklearn.datasets import load_boston
```

```python
x=load_boston()
data=pd.DataFrame(x.data,columns=x.feature_names)
y=x.target
```

```python
df=data
df
```

```python
#Creating Model
pipe=[]
pipe.append(('SC',StandardScaler()))
pipe.append(('PCA', PCA(n_components=8)))
pipe.append(('SGD',SGDRegressor()))

model=Pipeline(pipe)
CV_score=cross_val_score(model,data,y,cv=5)
score= '%s: %f (%f)'% ('SGDRegressor',CV_score.mean(),CV_score.std())
print (score)
```

```python
pipe=[]
pipe.append(('SC',StandardScaler()))
pipe.append(('PCA', PCA(n_components=8)))
pipe.append(('SGD',SGDRegressor()))

model=Pipeline(pipe)
CV_score=cross_val_score(model,data,y,cv=5)
print ('SGDRegressor:',CV_score.mean(),CV_score.std())
```

# Gridsearch CV

```python
from sklearn import datasets
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

dataset= datasets.load_diabetes()
alphavalue={'alpha':[1,0.01,0.001,0.0001,0]}

rd= Ridge()
grid= GridSearchCV(rd,alphavalue)

grid.fit (dataset.data,dataset.target)
print (grid)

print (grid.best_score_)
print (grid.best_params_)

```
