```python
df=pd.DataFrame (ds.data,columns=ds.feature_names)
df
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
mnb= MultinomialNB()
mnb.fit(x_train,y_train)
predmnb=mnb.predict(x_test)
print ('Model Score: ', mnb.score(x_train,y_train))
print ('Accuracy Score:',accuracy_score(y_test,predmnb))
print (confusion_matrix(y_test,predmnb))
print ('Classification Report:', classification_report(y_test,predmnb))


```

```python
svm= SVC()
svm.fit(x_train,y_train)
predsvm=svm.predict(x_test)
print ('Model Score: ', svm.score(x_train,y_train))
print ('Accuracy Score:',accuracy_score(y_test,predsvm))
print (confusion_matrix(y_test,predsvm))
print ('Classification Report:', classification_report(y_test,predsvm))
```

```python
dtc= DecisionTreeClassifier()
dtc.fit(x_train,y_train)
preddtc=dtc.predict(x_test)
print ('Model Score: ', dtc.score(x_train,y_train))
print ('Accuracy Score:',accuracy_score(y_test,preddtc))
print (confusion_matrix(y_test,preddtc))
print ('Classification Report:', classification_report(y_test,preddtc))
```

```python
knn= KNeighborsClassifier()
knn.fit(x_train,y_train)
predknn=knn.predict(x_test)
print ('Model Score: ', knn.score(x_train,y_train))
print ('Accuracy Score:',accuracy_score(y_test,predknn))
print (confusion_matrix(y_test,predknn))
print ('Classification Report:', classification_report(y_test,predknn))
```

```python
def svmkernel (ker):
    svm=SVC(kernel=ker)
    svm.fit(x_train,y_train)
    svm.score(x_train,y_train)
    predsvm=svm.predict(x_test)
    print (accuracy_score(y_test,predsvm))
    print (confusion_matrix(y_test,predsvm))
    print (classification_report(y_test,predsvm))
    
svmkernel ('rbf')   
```
