# T-TEST (Parametric Statistical Singnificance Test)

```python
from scipy import stats
```

```python
stats.ttest_1samp(df['FSIQ'],108) # P-value>0.05 so Ho(Null Hypothisis is Accepted) , The Mean of SFIQ is similar to 108
```

```python
stats.ttest_1samp(df['FSIQ'],88) # P-value>0.05 so Ho(Null Hypothisis) is Accepted , The Mean of SFIQ is similar to 88
```

```python
stats.ttest_ind(df['FSIQ'],df['PIQ']) #p-Value>0.5 Ho(Null Hypothisis) is Accepted
```

# ANOVA TEST

Analysis of Variance

```python
from scipy.stats import f_oneway
```

```python
Marks_SecationA= [70,40,60,78,90]
Marks_SecationB= [98,55,70,80,72]
Marks_SecationC= [70,30,86,59,96]
```

```python
f_oneway(Marks_SecationA,Marks_SecationB,Marks_SecationC)
```

```python
# p value is 0.82 which is greater thena 0.5 hence all group means are equal : null hypothisis is accepted 
```
