```python
irisds=pd.read_csv('https://raw.githubusercontent.com/dsrscientist/dataset1/master/iris.csv')

irisds
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')

df
```

```python
sns.violinplot(df,x='total_bill')
```

```python
sns.violinplot(df,x='tip')
```

```python
sns.stripplot(df,x='time',y='total_bill')
```

```python
sns.stripplot(df,x='day',y='total_bill')
```

```python
df['smoker'].value_counts()
```

```python
sns.histplot(df,x='total_bill',bins=20)
```

```python
sns.distplot(df['tip'],bins=10)
```

```python
sns.distplot(df['total_bill'])
```

```python
# kernel density estimation  parameter

sns.distplot(df['total_bill'],kde=False)
```

```python
sns.kdeplot(df['total_bill'])
```

```python
sns.jointplot(df,x='total_bill',y='tip')
```

```python
sns.boxplot(df,x='day',y='total_bill')
```

```python
sns.boxplot(df,x='day',y='total_bill',hue='smoker')
```

```python
sns.catplot(df,x='day',y='total_bill',hue='time',kind='bar')
```
